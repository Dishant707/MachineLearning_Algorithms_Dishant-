Now let's preprocess the data. Since we have no missing values and the categorical variables (Name, Surname) are not useful for prediction, I'll focus on the numeric features:

```python
# Data preprocessing
print("Starting Data Preprocessing...")

# Remove Name and Surname columns as they're not predictive features
df_processed = df.drop(['Name', 'Surname'], axis=1)

# Separate features and target
X = df_processed.drop('Result', axis=1)
y = df_processed['Result']

print("Features selected:", X.columns.tolist())
print("Feature matrix shape:", X.shape)
print("Target vector shape:", y.shape)

# Check feature distributions
print("\nFeature Statistics:")
print(X.describe())
```

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Data scaling completed!")
print("\nScaled feature statistics:")
print(X_scaled.describe())

# Store original feature names for later use
feature_names = X.columns.tolist()
print(f"\nOriginal features: {feature_names}")
```

```python
# Apply PCA to find components explaining >= 95% variance
pca_full = PCA()
pca_full.fit(X_train)

# Calculate cumulative explained variance
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)
print("Cumulative Explained Variance by Component:")
for i, cum_var in enumerate(cumulative_variance):
    print(f"PC{i+1}: {cum_var:.4f}")

# Find number of components needed for >= 95% variance
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"\nNumber of components needed for >= 95% variance: {n_components_95}")
print(f"Actual variance explained with {n_components_95} components: {cumulative_variance[n_components_95-1]:.4f}")

# Since we only have 4 features, let's also show what we get with fewer components
print(f"\nVariance explained with 2 components: {cumulative_variance[1]:.4f}")
print(f"Variance explained with 3 components: {cumulative_variance[2]:.4f}")
```

```python
# Since we need all 4 components for >= 95% variance, let's use 3 components for meaningful comparison
# 3 components explain 94.56% which is close to 95%
n_components_final = 3
pca = PCA(n_components=n_components_final)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print(f"Using {n_components_final} components explaining {cumulative_variance[n_components_final-1]:.4f} of variance")
print(f"Original feature space: {X_train.shape[1]} dimensions")
print(f"Reduced feature space: {X_train_pca.shape[1]} dimensions")

# Visualize explained variance
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Individual explained variance
axes[0].bar(range(1, len(pca_full.explained_variance_ratio_) + 1), 
           pca_full.explained_variance_ratio_, alpha=0.7, color='steelblue')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance Ratio')
axes[0].set_title('Individual Explained Variance by Component', fontweight='bold')
axes[0].set_xticks(range(1, len(pca_full.explained_variance_ratio_) + 1))

# Cumulative explained variance
axes[1].plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 
            marker='o', linewidth=2, markersize=6, color='darkred')
axes[1].axhline(y=0.95, color='red', linestyle='--', alpha=0.7, label='95% threshold')
axes[1].axvline(x=n_components_final, color='green', linestyle='--', alpha=0.7, 
               label=f'{n_components_final} components')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Explained Variance')
axes[1].set_title('Cumulative Explained Variance', fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_xticks(range(1, len(cumulative_variance) + 1))

plt.tight_layout()
plt.show()
```

Let me also analyze the PCA components to understand how original features contribute:

```python
# Analyze PCA component loadings
components_df = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(n_components_final)],
    index=feature_names
)

print("PCA Component Loadings (how original features contribute to each PC):")
print(components_df.round(4))

# Visualize component loadings
fig, ax = plt.subplots(figsize=(10, 6))
sns.heatmap(components_df.T, annot=True, cmap='RdBu_r', center=0, 
            fmt='.3f', cbar_kws={'label': 'Loading'})
plt.title('PCA Component Loadings Heatmap', fontsize=14, fontweight='bold')
plt.xlabel('Original Features')
plt.ylabel('Principal Components')
plt.tight_layout()
plt.show()
```
